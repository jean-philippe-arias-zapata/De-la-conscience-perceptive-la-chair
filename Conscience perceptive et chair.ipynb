### Composition des graphes

```python
Tot = nx.compose(Co, Ch)
```

```python
nx.draw_kamada_kawai(Tot, with_labels=True, font_weight='bold', node_color="c")
```

```python
Tot.add_edge("CoCoh", "ChCoh")
Tot.add_edge("CoCohInterSens", "ChCohInterSens")
Tot.add_edge("CoCohIntraSens", "ChCohIntraSens")

Tot.add_edge("CoCohInterSens", "CoCohIntraSens")
Tot.add_edge("ChCohIntraSens", "ChCohInterSens")

Tot.add_edge("CoCohInterMod", "ChCohInterMod")

Tot.add_edge("CoOuv", "ChOuv")
Tot.add_edge("CoEvid", "ChFoi")
Tot.add_edge("CoMa", "ChStyle")

Tot.add_edge("CoSpaSit", "ChSpaNeg")
Tot.add_edge("CoSpaSit", "ChTransi")
Tot.add_edge("CoSpaSit", "ChNonLoc")

Tot.add_edge("CoDyn", "ChLim")
Tot.add_edge("CoDyn", "ChOuvOnto")

Tot.add_edge("CoEqui", "ChAnalog")

Tot.add_edge("CoHabAnalog", "ChAnalog")
Tot.add_edge("CoHabAnalog", "ChGen")

Tot.add_edge("CoPui", "ChGenAnnex")
```

```python
nx.draw_kamada_kawai(Tot, with_labels=True, font_weight='bold')
```

```python
node_colors = []

for node in list(Tot.nodes):
    if node[:2] == "Co":
        node_colors.append("darkcyan")
    else:
        node_colors.append("olivedrab")
```

```python
nx.draw_kamada_kawai(Tot, with_labels=True, font_weight='bold', node_color=node_colors)
```

```python
edge_colors = []

for edge in list(Tot.edges()):
    if edge[0][:2] == "Co" and edge[1][:2] == "Ch":
        edge_colors.append("r")
    elif edge[0][:2] == "Ch" and edge[1][:2] == "Co":

        edge_colors.append("r")
    else:
        edge_colors.append("black")
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(ncols=1,nrows=1,figsize=(13, 12))
nx.draw_kamada_kawai(Tot, with_labels=True, font_weight='bold', node_color=node_colors, edge_color=edge_colors, font_size=8, ax=ax)

# Si vous souhaitez enregistrer le graphe, il suffit d'enlever le symbole # de la ligne suivante
#plt.savefig("conscience perceptive et chair.png", bbox_inches="tight")
```

```python
degree_dic = {}

for i in range(max(dict(Tot.degree).values()) + 1):
    a = []
    for key in dict(Tot.degree).keys():
        if Tot.degree[key] == i:
            a.append(key)
        degree_dic[i] = a
```

```python
important_degree = [degree_dic[i] for i in range(max(list(degree_dic.keys())) - 1, max(list(degree_dic.keys())) + 1)]
```

### Recherche de et nommer les communautés

```python
from networkx.algorithms.community.modularity_max import greedy_modularity_communities

communities = greedy_modularity_communities(Tot)
```

```python
communities_degree = []

for i in range(len(communities)):
    communities_degree.append({ele:Tot.degree[ele] for ele in communities[i]})
```

```python
communities_names = ["Pour autrui", "Ouverture", "Cohésions", "Analogicité", "Style", "Foi au monde"]

named_communities = {communities_names[i]:communities_degree[i] for i in range(len(communities_names))}
```

### Attributs et réduction du graphe

```python
reduced_named_communities = {}

for node in Tot.nodes():
    for item in named_communities.items():
        if node in item[1]:
            reduced_named_communities[node] = item[0]
            break
```

```python
nx.set_node_attributes(Tot, reduced_named_communities, "community")
```

```python
new_edges = []

for edge in nx.edges(Tot):
    new_edges.append((reduced_named_communities[edge[0]], reduced_named_communities[edge[1]]))
```

```python
reduced_graph = nx.Graph()
```

```python
reduced_graph.add_nodes_from(communities_names)
```

```python
reduced_graph.add_edges_from(new_edges)
```

```python
fig,ax = plt.subplots(ncols=1,nrows=1,figsize=(12, 9))
nx.draw_kamada_kawai(reduced_graph, with_labels=True, font_weight='bold', ax=ax)

# Si vous souhaitez enregistrer le graphe, il suffit d'enlever le symbole # de la ligne suivante
#plt.savefig("schéma corporel.png", bbox_inches="tight")
```

```python
A_reduced = nx.convert_matrix.to_pandas_adjacency(reduced_graph)
```
